# utilizando dataset de wines

```python
import pandas as pd
```

```python
df_rojos=pd.read_csv('winequality-red.csv',sep=";")
df_blancos=pd.read_csv('winequality-white.csv',sep=";")
```

```python
df_rojos.head()
```

```python
df_blancos.tail()
```

```python
df_rojos['color']='red'
df_blancos['color']="white"
```

```python
df_blancos.head()
```

```python
df_todos=pd.concat([df_rojos,df_blancos])
```

```python
df_todos.describe()
```

```python
df_todos['color'].value_counts()
```

```python
df_todos['quality'].value_counts()
```

```python
df_todos['color']=df_todos['color'].astype('category')
```

```python
df_todos.dtypes
```

```python
index=[x for x in range(df_todos.shape[0])]
df_todos['num']=index
df_todos.set_index('num',inplace=True)
```
