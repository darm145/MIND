# visualizacion

```python
import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_excel("wines.xlsx")
```

```python
df =pd.read_excel("wines.xlsx",index_col="num")
```

```python
df.head()
```

```python
df.color=pd.Categorical(df.color)
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
df["color_cod"]= df.color.cat.codes
```

```python
df.plot.scatter(x="fixed_acidity",y="volatile_acidity")
```

```python
df.plot.scatter(x="fixed_acidity",y="volatile_acidity", c="color_cod")
```

```python
df.plot.scatter(x="fixed_acidity",y="volatile_acidity", c="color_cod",colormap="inferno")
```

```python
plt.savefig("scatter1.png")
```

```python
#concentracion de puntos
df.plot.hexbin(x="fixed_acidity",y="volatile_acidity",gridsize=20)
```

```python
#concentracion de puntos
df.plot.hexbin(x="fixed_acidity",y="volatile_acidity",gridsize=20,colormap="inferno")
```

```python
plt.savefig("hex1.png")
```

```python
#hsitogramas(columna, clases)
df.hist(column="density",bins=100,grid=False, color="red", by ="color")
```

```python
df[["citric_acid","chlorides","density"]].plot.hist(bins=10,alpha=0.5)
```

```python
df.boxplot(column="residual_sugar",by="color",vert=False)
```
