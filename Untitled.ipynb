```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,4*np.pi,0.1)
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("sen(x) & cos(x)")
plt.legend(["sen(x)","cos(x)"])
plt.grid()
plt.title("seno y coseno de x")
```
